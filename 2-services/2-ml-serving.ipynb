## Project initialization

Create a project: a dedicated space where we can manage functions, artifacts and executions.

Use the ``username`` as param to create a personal project in the shared space.

```python
import digitalhub as dh
import os

project = dh.get_or_create_project(f"{os.environ['USER']}-mlops")
project
```

## Deploy a pre-trained model

We already have a model stored in the catalog, packed as MLFlow.


```python
model = project.get_model("mnist-classifier")
```

```python
model
```

```python
model.spec.parameters
```

We can download the asset.

```python
model.download()
```

We can define an mlflow serving function for exposing OpenInference v2 with our model.

```python
service = project.new_function("mlservice", kind="mlflowserve", model_name="mnist-classifier", path=model.key)
```

and run it

```python
run = service.run(action="serve", wait=False)
```

```python
run.status
```

With the service details we can invoke it: let's pick the URL and make a POST

```python
URL=run.status.service["url"]
```

```python
# invoke manually
import requests
import pprint
pp = pprint.PrettyPrinter(indent=2)

res = requests.post(f"http://{URL}",json={"text": f"{os.environ['USER']}-base"})
pp.pprint(res.json())
```

calling a service is a standard operation, we can leverage the SDK


```python
run.invoke(json={"text": f"{os.environ['USER']}-base"}).json()
```
